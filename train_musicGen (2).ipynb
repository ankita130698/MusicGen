```python
import os
import soundfile as sf
import torch

def generate_and_save_audio(wav_file_path, description_file_path, output_dir):
    # Read description with UTF-8 encoding
    with open(description_file_path, 'r', encoding='utf-8') as file:
        description = file.read().strip()

    # Generate audio
    res = model.generate([description], progress=True)

    # Convert the generated tensor to a NumPy array
    generated_audio = res[0].cpu().numpy()

    # Save audio
    audio_output_path = os.path.join(output_dir, os.path.basename(wav_file_path))
    sf.write(audio_output_path, generated_audio.T, 32000)  # Save with the same sample rate

    # Save description
    desc_output_path = os.path.join(output_dir, os.path.basename(description_file_path))
    with open(desc_output_path, 'w', encoding='utf-8') as file:
        file.write(description)

# Directories
music_data_dir = '/content/drive/MyDrive/TrainingData/extracted_1/music_data'
output_dir = '/content/drive/MyDrive/TestingData'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Process all files
files_processed = 0
for file_name in os.listdir(music_data_dir):
    if file_name.endswith('.wav'):
        wav_file_path = os.path.join(music_data_dir, file_name)
        description_file_path = wav_file_path.replace('.wav', '.txt')
        audio_output_path = os.path.join(output_dir, file_name)

        # Check if the generated file already exists
        if os.path.isfile(audio_output_path):
            print(f"File {audio_output_path} already exists. Skipping.")
            continue

        if os.path.isfile(description_file_path):
            generate_and_save_audio(wav_file_path, description_file_path, output_dir)
            files_processed += 1

print("Total Files Processed:", files_processed)
print("Generated audio and saved descriptions for new files in:", output_dir)
```

```python
from transformers import CLAPProcessor, CLAPModel
import torch
import torchaudio

# Load the processor and model
processor = CLAPProcessor.from_pretrained("openai/clap")
model = CLAPModel.from_pretrained("openai/clap")

def compute_clap_score(audio_file_path, text):
    # Load audio file
    waveform, sample_rate = torchaudio.load(audio_file_path)
    inputs = processor(audio=waveform, text=text, sampling_rate=sample_rate, return_tensors="pt")

    # Forward pass
    with torch.no_grad():
        outputs = model(inputs)

    # Compute CLAP score
    audio_embeds = outputs.audio_embeds
    text_embeds = outputs.text_embeds
    clap_score = torch.cosine_similarity(audio_embeds, text_embeds).item()

    return clap_score

# Example usage
audio_file_path = "path/to/your/generated/audio/file.wav"
text = "Your caption or description of the audio"
clap_score = compute_clap_score(audio_file_path, text)
print(f"CLAP Score: {clap_score}")
```
